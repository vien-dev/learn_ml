Import the neccessary libs

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

Create the first neural network with 1 layer, 1 node

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
```

Create the train data

```python
xt= np.array([-1.0, 0, 1.0, 2.0, 3.0, 4.0, 10.0], dtype=float)
yt= np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0, 19.0], dtype=float)
```

Train the model

```python
model.fit(xt, yt, epochs=500)
```

Now predict for a new value

```python
print(model.predict([11.0]))
```
